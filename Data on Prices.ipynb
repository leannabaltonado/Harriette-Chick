```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)   # Show all rows
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'left')

df = pd.read_csv("Dietary Food List - Nutrients.csv", header=None)

foods = df.iloc[0, 1:].tolist()

row_map = {
    25: ("Haiti", "Import"),
    26: ("Haiti", "Local"),
}
output = []
choices = ["Not Tracked", "No Specific Data"]

for row_idx, (country, type_) in row_map.items():
    # Row values for foods (skip first column)
    row_values = df.iloc[row_idx, 1:].tolist()
    for food, val in zip(foods, row_values):
        # Keep $ sign, fill missing
        if pd.isna(val) or str(val).strip() == '':
            price = np.random.choice(choices)
        else:
            price = str(val).strip()
        output.append({
            "Country": country,
            "Type": type_,
            "Food": food,
            "Price_USD": price
        })

tidy_prices = pd.DataFrame(output, columns=["Country", "Type", "Food", "Price_USD"])

country_order = ['Haiti']
type_order = ['Local', 'Import']
tidy_prices['Country'] = pd.Categorical(tidy_prices['Country'], categories=country_order, ordered=True)
tidy_prices['Type'] = pd.Categorical(tidy_prices['Type'], categories=type_order, ordered=True)

tidy_prices = tidy_prices.sort_values(['Country','Type','Food']).reset_index(drop=True)

print(tidy_prices)
```
